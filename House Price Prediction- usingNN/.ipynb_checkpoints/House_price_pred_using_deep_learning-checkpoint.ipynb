## 1. Exploratory Data Analysis(EDA)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('kc_house_data.csv')
data.head(7).T
```

```python
data.info()
```

```python
data.describe().transpose()
```

###### Data Visualization in EDA

```python
# Visualizing House Price
fig = plt.figure(figsize=(10,7))
fig.add_subplot(2,1,1)
sns.distplot(data['price'])
fig.add_subplot(2,1,2)
sns.boxplot(data['price'])
plt.tight_layout()
```

###### Learn -> 1. Most gharachi price 0-10 lacks madhe ahe.


```python
#visualizing square footage of (home,lot,above and basement)
fig = plt.figure(figsize=(16,5))
fig.add_subplot(2,2,1) # 2*2 grid, and 1st Subplot
sns.scatterplot(data['sqft_above'], data['price'])
fig.add_subplot(2,2,2)
sns.scatterplot(data['sqft_lot'],data['price'])
fig.add_subplot(2,2,3)
sns.scatterplot(data['sqft_living'],data['price'])
fig.add_subplot(2,2,4)
sns.scatterplot(data['sqft_basement'],data['price'])
```

```python
#visualizing bedrooms,bathrooms,floors,grade
fig = plt.figure(figsize=(15,7))
fig.add_subplot(2,3,1) # 2*3 grid (2Rows and 3Colms) & 1st Subplot
sns.countplot(data['bedrooms'])
fig.add_subplot(2,3,2)
sns.countplot(data['floors'])
fig.add_subplot(2,3,3)
sns.countplot(data['bathrooms'])
fig.add_subplot(2,3,4)
sns.countplot(data['grade'])
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(16,5))
sns.lineplot(data['floors'], data['price'])
```

```python
fig = plt.figure(figsize=(16,5))
sns.scatterplot(data['floors'], data['price'])
```
